**Analysis of relations among the happines factors**

Let see how 6 measures used to calculate happiness scores are interrelated with each other.

```python
plt.figure(figsize=(24,8))

plt.subplot(1,3,1)
plt.scatter( df_2020['Ladder score'], df_2020['Logged GDP per capita'])
plt.scatter( df_2021['Ladder score'], df_2021['Logged GDP per capita'])
plt.xlabel('Logged GDP per capita in 2020 and 2021')
plt.ylabel('Happiness score')
plt.legend(['2020','2021'])

plt.subplot(1,3,2)
plt.scatter( df_2020['Ladder score'], df_2020['Social support'])
plt.scatter( df_2021['Ladder score'], df_2021['Social support'])
plt.xlabel('Social support in 2020 and 2021')
plt.ylabel('Happiness score')
plt.legend(['2020','2021'])

plt.subplot(1,3,3)
plt.scatter( df_2020['Ladder score'], df_2020['Healthy life expectancy'])
plt.scatter( df_2021['Ladder score'], df_2021['Healthy life expectancy'])
plt.xlabel('Healthy life expectancy in 2020 and 2021')
plt.ylabel('Happiness score')
plt.legend(['2020','2021'])

plt.figure(figsize=(24,8))
plt.subplot(1,3,1)
plt.scatter( df_2020['Ladder score'], df_2020['Freedom to make life choices'])
plt.scatter( df_2021['Ladder score'], df_2021['Freedom to make life choices'])
plt.xlabel('Freedom to make life choices in 2020 and 2021')
plt.ylabel('Happiness score')
plt.legend(['2020','2021'])

plt.subplot(1,3,2)
plt.scatter( df_2020['Ladder score'], df_2020['Generosity'])
plt.scatter( df_2021['Ladder score'], df_2021['Generosity'])
plt.xlabel('Generosity in 2020 and 2021')
plt.ylabel('Happiness score')
plt.legend(['2020','2021'])
plt.subplot(1,3,3)

plt.scatter( df_2020['Ladder score'], df_2020['Perceptions of corruption'])
plt.scatter( df_2021['Ladder score'],df_2021['Perceptions of corruption'])
plt.xlabel('Perceptions of corruption in 2020 and 2021')
plt.ylabel('Happiness score')
plt.legend(['2020','2021'])
```

Healthy life expectancy, GDP per capita and social support are main correlations. 

```python
correlations_2020 = df_2020[['Logged GDP per capita', 'Social support', 'Healthy life expectancy',
                                 'Freedom to make life choices', 'Generosity','Perceptions of corruption',
                                 'Dystopia + residual']]
sns.pairplot(correlations_2020, kind='reg')
```
